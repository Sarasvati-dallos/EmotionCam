```python
# CELDA 1 — Instalar dependencias
%%bash
sudo apt-get update -qq
sudo apt-get install -y python3-pip build-essential git zip unzip \
    openjdk-17-jdk autoconf libtool pkg-config zlib1g-dev \
    libncurses5-dev libncursesw5-dev cmake libffi-dev libssl-dev -qq
pip install -q buildozer cython
echo '✅ Listo'
```

```python
# CELDA 2 — Subir emotion_model.tflite y emotion_labels.json
from google.colab import files
print('Sube: emotion_model.tflite y emotion_labels.json')
uploaded = files.upload()
print('Subidos:', list(uploaded.keys()))
```

```python
# CELDA 3 — Clonar repo y preparar archivos
import os, shutil
os.system('git clone https://github.com/Sarasvati-dallos/EmotionCam.git /content/EmotionCam')
for f in ['emotion_model.tflite', 'emotion_labels.json']:
    if os.path.exists(f):
        shutil.copy(f, f'/content/EmotionCam/{f}')
        print(f'✅ {f} copiado')
    else:
        print(f'❌ {f} no encontrado')
print('\nArchivos:', os.listdir('/content/EmotionCam'))
```

```python
# CELDA 4 — Verificar
import os
required = ['main.py','buildozer.spec','modules/database.py','emotion_model.tflite','emotion_labels.json']
all_ok = True
for f in required:
    p = f'/content/EmotionCam/{f}'
    ok = os.path.exists(p)
    print(f"{'✅' if ok else '❌'} {f}")
    if not ok: all_ok = False
print('\n✅ Listo para compilar' if all_ok else '\n❌ Faltan archivos')
```

```python
# CELDA 5 — COMPILAR (30-40 min, no cierres el navegador)
import subprocess
print('Compilando... ☕ espera')
result = subprocess.run(
    ['buildozer', 'android', 'debug'],
    input='y\ny\ny\n',
    capture_output=True, text=True,
    cwd='/content/EmotionCam'
)
print(result.stdout[-3000:])
print(result.stderr[-2000:])
```

```python
# CELDA 6 — Descargar APK
import os, glob
from google.colab import files
apks = glob.glob('/content/EmotionCam/bin/*.apk')
if apks:
    print(f'✅ APK listo: {apks[0]}')
    files.download(apks[0])
    print('\n📱 Instala en Android:')
    print('1. Pasa el APK al teléfono')
    print('2. Activa fuentes desconocidas en Ajustes')
    print('3. Abre el APK e instala')
else:
    print('❌ APK no encontrado')
    print(result.stderr[-3000:])
```
